Einfaches Beispiel für Client-Server-Interaktion über Sockets.

```python
# notwendige imports
import socket
import threading
import time

HOST = "127.0.0.1"
PORT = 5000
```

Server definieren.
Zeigt empfangene Daten, hängt " *" an und sendet diese an Client zurück.

```python
def run_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind((HOST, PORT))
    server_socket.listen(1)

    print("🟢 Server läuft und wartet auf Verbindung...")

    conn, addr = server_socket.accept()  # blocking
    print(f"🔗 Verbunden mit {addr}")

    while True:
        data = conn.recv(1024)  # blocking

        if not data:
            print("🔴 Client hat Verbindung beendet")
            break

        print("📥 Server empfangen:", data.decode())

        conn.sendall(data + b" *")

    conn.close()
    server_socket.close()
    print("🛑 Server beendet")
```

Server startet im Hintergrund als Thread.

```python
# Server startet als Thread im Hintergrund, wir können CLient und Server
# nicht in einer gemeinsamen Zelle ausführen.
server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

# kleine Pause, damit Server bereit ist
time.sleep(1)
```

Client ausführen.
Sendet Nachricht an Server und empfängt Echo.
Zeigt beides an.

```python
def run_client():
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    print("🔵 Client verbindet sich...")
    client_socket.connect((HOST, PORT))  # blocking

    message = "Hello World"
    print("📤 Client sendet:", message)

    client_socket.sendall(message.encode())

    data = client_socket.recv(1024)
    print("📥 Client empfängt:", data.decode())

    client_socket.close()
    print("🛑 Client beendet")
```

Client ausführen

```python
run_client()
```
