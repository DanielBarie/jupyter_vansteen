# MPI-Demo mit Python (`mpi4py`)

Dieses Notebook zeigt ein minimales MPI-Beispiel mit Python.

## Ziel
Jeder Prozess gibt seine eigene `rank` sowie die Gesamtzahl der Prozesse `size` aus.

## Hinweis
MPI-Programme startet man **nicht normal innerhalb einer einzelnen Notebook-Zelle mit Parallelität**, sondern typischerweise über die Kommandozeile, zum Beispiel:

```bash
mpiexec -n 4 python hello_mpi.py
```

Damit das Ganze trotzdem als Notebook nutzbar ist, enthält dieses Notebook:
1. die Erklärung,
2. den MPI-Code,
3. eine Zelle, die daraus automatisch eine Python-Datei erzeugt,
4. und die passenden Startbefehle.

````python
from mpi4py import MPI
````

## Grundidee

- `COMM_WORLD` ist der Standard-Kommunikator.
- `Get_size()` liefert die Anzahl aller Prozesse.
- `Get_rank()` liefert die ID des aktuellen Prozesses.


````python
# Initialize MPI
comm = MPI.COMM_WORLD
size = comm.Get_size()
rank = comm.Get_rank()

print(f"Hello from process {rank} out of {size}")

# In mpi4py wird Finalize beim Programmende in der Regel automatisch behandelt
````

## Dasselbe als Skript speichern

Die nächste Zelle schreibt den Beispielcode in eine Datei `hello_mpi.py`.
Danach kann diese mit `mpiexec` ausgeführt werden.


````python
code = '''from mpi4py import MPI

comm = MPI.COMM_WORLD
size = comm.Get_size()
rank = comm.Get_rank()

print(f"Hello from process {rank} out of {size}")
'''

with open("hello_mpi.py", "w", encoding="utf-8") as f:
    f.write(code)

print("Datei hello_mpi.py wurde erstellt.")
````

## Ausführung

Starten des Skripts in einer Shell oder einem Terminal mit:

```bash
mpiexec -n 4 python hello_mpi.py
```

Mögliche Ausgabe:

```text
Hello from process 0 out of 4
Hello from process 1 out of 4
Hello from process 2 out of 4
Hello from process 3 out of 4
```

Die Reihenfolge der Ausgaben kann variieren, weil die Prozesse parallel laufen.


## Optional: kleine Erweiterung

Man kann zusätzlich demonstrieren, dass nur Prozess 0 eine Sammelmeldung ausgibt.


````python
comm = MPI.COMM_WORLD
size = comm.Get_size()
rank = comm.Get_rank()

local_message = f"Ich bin Prozess {rank}"
all_messages = comm.gather(local_message, root=0)

if rank == 0:
    print("Gesammelte Nachrichten:")
    for msg in all_messages:
        print("-", msg)
````

Den erweitertenn Code z. B. als `gather_demo.py`speichern und wieder mit `mpiexec` starten:
